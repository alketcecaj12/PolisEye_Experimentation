```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import datetime
```

```python
# carica dati 
data = pd.read_csv('../data/Modena_20170416_20170817', 
                  header = None, names=['date', 'cell_num', 'nr_people'], index_col = None)
```

```python
data
```

```python
data.describe()
```

```python
# funzione che crea un campo 'date' nel formato yyyy-mm-dd 
li = []
hours_list = []
minutes_list = []

def sistema_data(l, hours_list, minutes_list):
    for i in data['date']: 
        date_ymd = i[:16]
        year = date_ymd[:4]
        month = date_ymd[4:6]
        day = date_ymd[6:8]
        hour = date_ymd[9:11]
        #print(hour)
        hours_list.append(int(hour))
        minutes = date_ymd[11:13]
        minutes_list.append(int(minutes))
                            
        ymd_string = year+'-'+month+'-'+day
        l.append(ymd_string)
     
        
    return [l, hours_list, minutes_list]
    
l2 = sistema_data(li, hours_list, minutes_list)

data['date'] = l2[0]
data['hours'] = l2[1]
data['minutes']= l2[2]
```

```python
data.describe()
```

```python
data['nr_people'] = data['nr_people'].replace(65535, np.nan)
print(data.describe())
#data = data.replace(65534, np.nan)
#print(data.shape)
```

```python
data.info()
```

```python
agg_date_cell = data.groupby(['date', 'cell_num']) 
count = 0
for i , k in agg_date_cell:
    count +=1
    if count > 9:
        count = 0
        break
    k['nr_people'].plot(figsize = (14, 6), title='Cell '+str(i[1])+', Date '+str(i[0]));
    plt.show()
```

```python
# agg. per numero cella e giorno
cell_date = data.groupby(by = ['cell_num'])
```

```python
count = 0
for i, k in cell_date:
    print(i, k)
    count+=1
    if count > 3:
        break
```

```python
new_data = pd.DataFrame({"date": [], 
                         "cell_num": [],
                         "nr_people": [], 
                         "hours":[],
                         "minutes":[]})
count = 0
for i, k in cell_date:
    k = k.fillna(method = 'ffill')
    k = k.fillna(method = 'bfill')
    new_data= pd.concat([new_data, k])
    del k
    count +=1
    if count % 100 == 0:
        print(count) 
```
