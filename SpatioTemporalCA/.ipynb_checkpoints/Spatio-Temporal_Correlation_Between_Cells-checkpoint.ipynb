### Heatmap a parire dalla matrice di correlazione - per tutto il periodo

```python
agg_by_cell_day = data.groupby(by = ['date','cell_num'])
```

```python
count = 0
dict2data1 = {}
dict2data2 = {}
for i, k in agg_by_cell_day:
    count += 1 
    dict2data1[i[1]] = k['nr_people'].tolist()
    if count % 13 == 0:
        dict2data2[count] = dict2data1 
        dict2data1 = {}   
```

```python
rg = [13, 26, 39, 52, 65, 78, 91, 104, 117, 130, 143, 156, 169, 182, 195, 208, 221 ]

for i in rg:
    data4heatmap = pd.DataFrame.from_dict(dict2data2[i])
    #print(dict2data2.get(key))
    #%matplotlib qt 
    fig, ax = plt.subplots(figsize=(30,30)) 
    corr = data4heatmap.corr() 
    ax = sns.heatmap(
        corr, 
        vmin=-1, vmax=1, center=0,
        cmap=sns.diverging_palette(20, 220, n=200),
        square=False,
        cbar=False
    )
    ax.set_xticklabels(
        ax.get_xticklabels(),
        rotation=90,
        horizontalalignment='right',
        fontsize=22
    );

    ax.set_yticklabels(
        ax.get_yticklabels(),
        rotation=0,
        horizontalalignment='right',
        fontsize=22
    );
```

### Heatmap di dati (celle) con aggregazione giornaliera 

```python
agg_by_cell_day = data.groupby(by = ['date','cell_num'])
```

```python
dict2data_day = {}
count = 0
for i,k in agg_by_cell_day:
    count +=1
    if count >221: break
    dict2data_day[i] = k['nr_people'].tolist()
```

```python
data4_daily_heatmap = pd.DataFrame.from_dict(dict2data_day)
```

```python
%matplotlib qt 
fig, ax = plt.subplots(figsize=(30,30)) 
corr = data4_daily_heatmap.corr()
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=False,
    cbar=False
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=90,
    horizontalalignment='right'
);

ax.set_yticklabels(
    ax.get_yticklabels(),
    rotation=0,
    horizontalalignment='right'
);
```
