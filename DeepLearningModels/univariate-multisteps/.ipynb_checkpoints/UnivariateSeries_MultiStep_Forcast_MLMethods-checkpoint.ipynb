## MLP per i dati di una sola cella - previsione a piÃ¹ di 1 step   

### Crea finestra di lag

```python
# split a univariate sequence into samples
def split_sequence(sequence, n_steps_in, n_steps_out):
    X, y = list(), list()
    for i in range(len(sequence)):
         # find the end of this pattern
         end_ix = i + n_steps_in
         out_end_ix = end_ix + n_steps_out
         # check if we are beyond the sequence
         if out_end_ix > len(sequence):
             break
         # gather input and output parts of the pattern
         seq_x, seq_y = sequence[i:end_ix], sequence[end_ix:out_end_ix]
         X.append(seq_x)
         y.append(seq_y)
    return np.array(X), np.array(y)

# choose a number of time steps
n_steps_in, n_steps_out = 96, 96

# split into samples
X, y = split_sequence(series, n_steps_in, n_steps_out)
count = 0
for i in range(len(X)):
    print(X[i], y[i])
    count +=1
    if count>5: break
```

### Normalizza i dati 

```python
# normalizza 
scaler = MinMaxScaler()
    
#fit scaler on data
#X = X.reshape(-1, 1)
scaler.fit(X)
    
#apply transform
X = scaler.transform(X)
print(X[0:10])

# fit scaler on data
#y = y.reshape(-1, 1)
scaler.fit(y)

# apply transform
y = scaler.transform(y)
print(y[0:10])
```

### Dividi i dati in train e test

```python
# split into train and test sets
train_size = int(len(X) * 0.67)
test_size = len(X) - train_size
train_X, test_X = X[0:train_size,:], X[train_size:len(X),:]

train_y, test_y = y[0:train_size], y[train_size:len(y)]
print('X', len(train_X), len(test_X))
print('y ', len(train_y), len(test_y))
print(test_X.shape)
print(test_y.shape)

print(train_X.shape)
print(train_y.shape)
```

### Definisci modello 

```python
# define model
model = Sequential()
model.add(Dense(100, activation='relu', input_dim=n_steps_in)) 
model.add(Dense(n_steps_out))

#opt = SGD(lr=0.01, momentum=0.9)
model.compile(optimizer='adam', loss='mse')

# fit model
model.fit(train_X, train_y, batch_size=64, epochs=1500, verbose=2)
```

### Trova il numero ottimale di neuroni. 
<p> L'ultimo calcolo fatto per questo punto ha mostrato che il numero ottimale per i neuroni Ã¨ 1000. Questo numero puÃ² cambiare se cambia la dimensione della finestra e i step di previsione</p>

```python
# fit model with given number of nodes, returns test set accuracy
def evaluate_model(n_nodes, batch_size_i, train_X, train_y, val_X, val_y):
    
    # define model
    model = Sequential()
    model.add(Dense(n_nodes, activation='relu', input_dim=n_steps_in)) 
    model.add(Dense(n_steps_out))

    model.compile(loss='mse',optimizer='adam')
   
    # test_X[0:672]
    history = model.fit(train_X, train_y, batch_size=batch_size_i, epochs=10, verbose=0)
    
    test_acc = model.evaluate(val_X, val_y, verbose=0)
    return history, test_acc  

plt.figure(figsize = (16, 6))
num_nodes = [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000, 
             1500, 2000, 2500, 3000, 4000, 5000, 6000, 7000, 8000]
batch_sizes = [32, 64, 96, 128, 192, 224, 384, 672]

for batch_i in batch_sizes:
    print('***---------------------------- batch size = ', batch_i)
    for n_nodes in num_nodes:
        # evaluate model with a given number of nodes
        history, result = evaluate_model(n_nodes, batch_i, train_X, train_y, test_X, test_y)
        
        # summarize final test set accuracy
        print('number of nodes=%d: %.3f' % (n_nodes, result))
         
        # plot learning curve
        plt.plot(history.history['loss'], label=str(n_nodes))       
    
    # show the plot
    plt.legend(prop={'size': 12})
    plt.show()
    
```

### Il tuning degli iperparametri

<p>
Gli iperparametri sui quali  fare tuning sono :<br />
    
1- Batch size (tipicamente uguale a 32 oppure 64)<br />
    
2- Numero di epoch (bisogna metterlo molto grande per poi usare early stopping)<br/>

3- Numero dei neuroni / layers (forse non ha senso un numero di neuroni piÃ¹ alto di mille (dipende da dim finestra)) <br/> 

4- Learning rate  (SGD(lr = 0.01) potrebbe aiutare nel caso di MLP) <br/> 

5- Formatto dati (dimesione della finestra)
</p>

```python
def evaluate_model(n_nodes, n_epochs, batch_size_i, train_X, train_y, val_X, val_y):
    
    # define model
    model = Sequential()
    model.add(Dense(n_nodes, activation='relu', input_dim=n_steps_in)) 
    model.add(Dense(n_steps_out))

    model.compile(loss='mse',optimizer='adam')
   
    # test_X[0:672]
    history = model.fit(train_X, train_y, batch_size=batch_size_i, epochs=n_epochs, verbose=0)
    
    test_acc = model.evaluate(val_X, val_y, verbose=0)
    return history, test_acc  



# hyperparameters --> batch_size, num_nodes, n_epochs, learning rate, dati input(dimensione finestra)

opt = SGD(lr=0.01, momentum=0.9)
 # mettere epoch numero molto grande e poi usare l'early stopping
num_nodes = [100, 300, 500, 600, 700, 800, 900, 1000]  # fino a 500 dovrebbero bastare 
                                                                 # non ha senso mettere un numero molto grande
batch_sizes = [32, 64, 128]


for batch_i in batch_sizes:
    print('***---------------------------- batch size = ', batch_i)
    for n_nodes in num_nodes:
        # evaluate model with a given number of nodes
        history, result = evaluate_model(n_nodes, 1000, batch_i, train_X, train_y, test_X, test_y)
        
        # summarize final test set accuracy
        print('loss = %.3f  --> number of nodes = %d: '%(result, n_nodes)+'epochs = '
              +str(1000)+',batch ='+str(batch_i))
```

### Calcolo dell'errore per tutte le celle con previsione multistep  

```python
# choose a number of time steps for the window
from sklearn.preprocessing import MinMaxScaler
count = 0
error_per_cell = {}
all_cell_difference = np.array([])
n_steps_in, n_steps_out = 672, 96

for cell, df_i in agg_by_cell:
    print(cell)
    series = df_i['nr_people'].values
    
    # split into samples
    X, y = split_sequence(series, n_steps_in, n_steps_out)
    
    # normalizza 
    scaler = MinMaxScaler()
    
    #fit scaler on data
    #X = X.reshape(-1, 1)
    scaler.fit(X)
    
    #apply transform
    X = scaler.transform(X)

    # fit scaler on data
    #y = y.reshape(-1, 1)
    scaler.fit(y)

    # apply transform
    y = scaler.transform(y)
    
    # split into train and test sets
    train_size = int(len(X) * 0.67)
    test_size = len(X) - train_size
    train_X, test_X = X[0:train_size,:], X[train_size:len(X),:]
    train_y, test_y = y[0:train_size], y[train_size:len(y)]
    
    # define model
    model = Sequential()
    model.add(Dense(600, activation='relu', input_dim=n_steps_in)) 
    model.add(Dense(n_steps_out))
    model.compile(optimizer='adam', loss='mse')
    
    # fit model
    model.fit(train_X, train_y, epochs=500, batch_size=64, verbose=0)
   
    # predict and compute error
    expected = test_y
    predicted = []
    
    # show boxplot
    plt.figure(figsize = (14,2))
    plt.boxplot(series, vert= False);
    plt.title('Error distrib. for cell '+str(cell))    
    plt.show()
    
    
    for i in range(len(test_X)):
        x_input = test_X[i].reshape((1, n_steps_in))
        yhat = model.predict(x_input, verbose=0)
        #predicted.append(np.rint(yhat[0]))
        predicted.append(np.around(yhat[0], decimals=1))
    
    # abs difference
    difference = abs((expected - predicted))
    #difference =  np.reshape(difference, difference.shape[0] * difference.shape[1])
    
    # analytical data
    all_cell_difference = np.append(all_cell_difference, difference)
    error_per_cell[cell] = difference
    
    # show boxplot
    plt.figure(figsize = (14,18))
    plt.boxplot(difference, vert= False);
    plt.title('Error distrib. for cell '+str(cell))    
    plt.show()
    print('Mean Absolute Error = ', np.mean(difference))
    
    # controll
    count +=1
    print('iteration = ', count)
```

```python
print('Mean Absolute Error = ', np.mean(all_cell_difference))
plt.figure(figsize = (14,2))
plt.boxplot(all_cell_difference, vert= False);
plt.title('Error distrib. for for all cells')    
plt.show()
```

### Convolutional NN per i dati di una sola cella - previsione piÃ¹ di 1 step temporale 

```python
# split a univariate sequence into samples
def split_sequence(sequence, n_steps_in, n_steps_out):
    X, y = list(), list()
    for i in range(len(sequence)):
        # find the end of this pattern
        end_ix = i + n_steps_in
        out_end_ix = end_ix + n_steps_out
        # check if we are beyond the sequence
        if out_end_ix > len(sequence):
            break
        # gather input and output parts of the pattern
        seq_x, seq_y = sequence[i:end_ix], sequence[end_ix:out_end_ix]
        X.append(seq_x)
        y.append(seq_y)
    return np.array(X), np.array(y)
```

```python
# choose a number of time steps
n_steps_in, n_steps_out = 3, 24

# split into samples
X, y = split_sequence(series, n_steps_in, n_steps_out)

# reshape from [samples, timesteps] into [samples, timesteps, features]
n_features = 1
X = X.reshape((X.shape[0], X.shape[1], n_features))
```

```python
# define model
model = Sequential()
model.add(Conv1D(filters=64, kernel_size=2, activation='relu', input_shape=(n_steps_in, n_features)))
model.add(MaxPooling1D(pool_size=2)) 
model.add(Flatten())
model.add(Dense(50, activation='relu')) 
model.add(Dense(n_steps_out)) 
model.compile(optimizer='adam', loss='mse')

# fit model
model.fit(X, y, epochs=500, verbose=2)
```

```python
# demonstrate prediction
x_input = np.array([25, 24, 23])
x_input = x_input.reshape((1, n_steps_in, n_features)) 
yhat = model.predict(x_input, verbose=0)
print(yhat)
```

### LSTM per i dati di una sola cella -  previsione piÃ¹ di 1 step temporale

```python
# split a univariate sequence into samples
def split_sequence(sequence, n_steps_in, n_steps_out):
    X, y = list(), list()
    for i in range(len(sequence)):
        # find the end of this pattern
        end_ix = i + n_steps_in
        out_end_ix = end_ix + n_steps_out
        # check if we are beyond the sequence
        if out_end_ix > len(sequence):
            break
        # gather input and output parts of the pattern
        seq_x, seq_y = sequence[i:end_ix], sequence[end_ix:out_end_ix]
        X.append(seq_x)
        y.append(seq_y)
    return np.array(X), np.array(y)

# choose a number of time steps
n_steps_in, n_steps_out = 3, 19

# split into samples
X, y = split_sequence(series, n_steps_in, n_steps_out)

# reshape from [samples, timesteps] into [samples, timesteps, features]
n_features = 1
X = X.reshape((X.shape[0], X.shape[1], n_features))
```

```python
# define model
model = Sequential()
model.add(LSTM(100, activation='relu', return_sequences=True, input_shape=(n_steps_in,n_features)))
model.add(LSTM(100, activation='relu')) 
model.add(Dense(n_steps_out)) 
model.compile(optimizer='adam', loss='mse')

# fit model
model.fit(X, y, epochs=50, verbose=2)
```

```python
# demonstrate prediction
x_input = np.array([25, 24, 23])
x_input = x_input.reshape((1, n_steps_in, n_features)) 
yhat = model.predict(x_input, verbose=2)
print(yhat)
```
