```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
import time
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers.convolutional import Conv1D
from keras.layers.convolutional import MaxPooling1D
from keras.optimizers import SGD
from sklearn.preprocessing import MinMaxScaler
from keras import metrics

data = pd.read_csv('/Users/alket/Desktop/dati/new_data_backfill_forwfill.csv',index_col = 0)

agg_by_cell = data.groupby(by = ['cell_num'])
```

```python
serie_dati = agg_by_cell.get_group('496-1255')
serie_dati = serie_dati.iloc[::4, :]
print(len(serie_dati))
serie_dati = serie_dati['nr_people'].values
print(serie_dati[:12])
```

```python
plt.figure(figsize = (18,4))
plt.plot(serie_dati[880:1560])
plt.show()
```

```python
# split a univariate sequence into samples
def split_sequence(sequence, n_steps_in, n_steps_out):
    X, y = list(), list()
    for i in range(len(sequence)):
         # find the end of this pattern
         end_ix = i + n_steps_in
         out_end_ix = end_ix + n_steps_out
         # check if we are beyond the sequence
         if out_end_ix > len(sequence):
             break
         # gather input and output parts of the pattern
         seq_x, seq_y = sequence[i:end_ix], sequence[end_ix:out_end_ix]
         X.append(seq_x)
         y.append(seq_y)
    return np.array(X), np.array(y)

# choose a number of time steps
n_steps_in, n_steps_out = 24,24

# split into samples
X, y = split_sequence(serie_dati, n_steps_in, n_steps_out)
count = 0
for i in range(len(X)):
    print(X[i], y[i])
    count +=1
    if count>2: break
```

```python
train_size = 2500


train_X, test_X = X[:train_size], X[train_size:]
train_y, test_y = y[:train_size], y[train_size:]
print(train_X.shape, test_X.shape)
```

```python
# define model
model = Sequential()
model.add(Dense(500, activation='relu', input_dim=n_steps_in)) 
model.add(Dense(n_steps_out))

#opt = SGD(lr=0.01, momentum=0.9)
model.compile(optimizer='adam', loss='mse' , metrics=[metrics.mae, 'accuracy'])

# fit model
model.fit(train_X, train_y, batch_size=64, epochs=50, verbose=2)
```

```python
# predict and compute error
expected = test_y
predicted = []
for i in range(len(test_X)):
    x_input = test_X[i].reshape((1, n_steps_in)) # + 1 per la nuova feature aggiunta
    yhat = model.predict(x_input, verbose=0)
    #predicted.append(np.rint(yhat[0]))   
    predicted.append(np.around(yhat[0], decimals=1)) 
```

```python
difference = abs(expected - predicted)
print(type(difference))
print(difference.shape[0], difference.shape[1])

# comment line below to see all the boxplots for every number of steps from 1 to n_steps_out
#difference =  np.reshape(difference, difference.shape[0] * difference.shape[1])

plt.figure(figsize = (19,4))
plt.boxplot(difference, vert= True);
plt.show()

#plt.figure(figsize = (14, 9))
#plt.boxplot(one_cell['nr_people'].values, vert = False)
#plt.boxplot(difference, vert= False);
#plt.xticks( rotation='vertical')
#plt.show()
```
