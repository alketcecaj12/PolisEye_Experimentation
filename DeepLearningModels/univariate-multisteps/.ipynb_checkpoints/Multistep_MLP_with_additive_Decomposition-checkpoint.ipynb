### Multistep_MLP_with_additive_Decomposition

```python
from statsmodels.compat.pandas import deprecate_kwarg
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
# additive decompose a contrived additive time series
from random import randrange
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose 
```

```python
data = pd.read_csv('/Users/alket/Desktop/dati/new_data_backfill_forwfill.csv', index_col = 0, 
                   header=0, parse_dates=True)
```

```python
agg_by_cell = data.groupby(by = ['cell_num'])
print(len(agg_by_cell), 'celle')
```

```python
serie_dati = agg_by_cell.get_group('495-1266')
serie_dati 
```

```python
counter = 0
dict2data = {}
error_list = []

print(counter)
counter +=1
dates4dec = []
cell_values = []

for index, row in serie_dati.iterrows():
    
        date = row['date']
        h = str(row['hours'])
   
        h = h.split('.')
    
        if len(h[0])<2:
            h = h[1]+h[0]
        else: 
            h = h[0]
   
        minutes = str(row['minutes'])
        m = ''
        minutes = minutes.split('.')
        if len(minutes[0])<2: 
            m = minutes[0] +'0'
        else: 
            m = minutes[0]
        #print(date, h, m)
        data_f = date+' '+h+':'+m+':'+'00'
        #print(data_f)
        cell_values.append(row['nr_people'])
        dates4dec.append(data_f) 
```

```python
dict_i = {'ds': dates4dec, 'y':cell_values}
data4deco = pd.DataFrame(dict_i, index=None, columns=None)  
data4deco.head()
```

```python
#data4deco.reset_index(inplace=True)
#data4deco['ds'] = pd.to_datetime(data4deco['ds'])
#data4deco = data4deco.set_index('ds')
#data4deco.head()
```

```python
data4deco.info()
data4deco.y.interpolate(inplace=True)
```

```python
decompfreq = 24 * 4
result = seasonal_decompose(data4deco['y'], freq=decompfreq, model='additive') 
with plt.rc_context():
     plt.rc("figure", figsize=(18,9))
     result.plot()
     plt.show()
```
