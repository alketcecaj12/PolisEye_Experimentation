### Creazione di features 

#### Crea feature 'is_buisness_day' che vale 1 se Ã¨ un giorno lavorativo, 0 altrimenti

```python
# crea un campo 'is_business_day' con valore binario
def find_business_days(data):
    business_day = []
    for i in data['date']:
        #print(i)
        s = i.split('-')
        year = int(s[0])
       # print(year, '----------------_>>>>><')
        month = int(s[1])
        day = int(s[2])
        ##print(day, ' ', month, ' ', year)
    
        dayofweek = datetime.date(year, month, day).strftime("%A")
        if ((dayofweek == 'Saturday') | (dayofweek == 'Sunday')): 
            business_day.append(0)
        else: 
            business_day.append(1)
     
    return business_day

bd = find_business_days(series)
series['is_business_day']= bd  
```

#### Crea feature 'is_April', 'is_May' ecc.. che vale 1 se si tratta del mese in questione, vale 0 altrimenti

```python
# crea un campo 'is_Month' con valore binario
def find_month(data):
    april = []
    may = []
    june = []
    july = []
    august = []
    
    for i in data['date']:
        #print(i)
        s = i.split('-')
        year = int(s[0])
       # print(year, '----------------_>>>>><')
        month = int(s[1])
        #print(month)
        if month == 4:
            april.append(1)
            may.append(0)
            june.append(0)
            july.append(0)
            august.append(0)
        if month == 5:
            april.append(0)
            may.append(1)
            june.append(0)
            july.append(0)
            august.append(0)
        if month == 6:
            april.append(0)
            may.append(0)
            june.append(1)
            july.append(0)
            august.append(0)
        if month == 7:
            april.append(0)
            may.append(0)
            june.append(0)
            july.append(1)
            august.append(0) 
        if month == 8:
            april.append(0)
            may.append(0)
            june.append(0)
            july.append(0)
            august.append(1)     
     
    return april, may, june, july, august

april, may, june, july, august = find_month(series)
series['is_April']= april
series['is_May']= may
series['is_June']= june
series['is_July']= july
series['is_August']= august
```

#### Crea feature is_Day es. 'is_Monday' che vale 1 se si tratta del giorno in questione, vale 0 altrimenti

```python
# crea un campo 'is_Day' con valore binario
def find_month(data):
    monday = []
    tuesday = []
    wednesday = []
    thursday = []
    friday = []
    
    for i in data['date']:
        s = i.split('-')
        year = int(s[0])
        month = int(s[1])
        day = int(s[2])
        
        dayofweek = datetime.date(year, month, day).strftime("%A")
        if (dayofweek == 'Monday'): 
            monday.append(1)
            tuesday.append(0)
            wednesday.append(0)
            thursday.append(0)
            friday.append(0)
        if (dayofweek == 'Tuesday'):
            monday.append(0)
            tuesday.append(1)
            wednesday.append(0)
            thursday.append(0)
            friday.append(0)
        if (dayofweek == 'Wednesday'):
            monday.append(0)
            tuesday.append(0)
            wednesday.append(1)
            thursday.append(0)
            friday.append(0)
        if (dayofweek == 'Thursday'):
            monday.append(0)
            tuesday.append(0)
            wednesday.append(0)
            thursday.append(1)
            friday.append(0) 
        if (dayofweek == 'Friday'):
            monday.append(0)
            tuesday.append(0)
            wednesday.append(0)
            thursday.append(0)
            friday.append(1)     
        if (dayofweek == 'Saturday'):
            monday.append(0)
            tuesday.append(0)
            wednesday.append(0)
            thursday.append(0)
            friday.append(0)          
        if (dayofweek == 'Sunday'):
            monday.append(0)
            tuesday.append(0)
            wednesday.append(0)
            thursday.append(0)
            friday.append(0) 
              
    return monday, tuesday, wednesday, thursday, friday

monday, tuesday, wednesday, thursday, friday = find_month(series)
print(len(monday))
print(len(tuesday))
print(len(friday))
series['is_Monday']= monday
series['is_Tuesday']= tuesday
series['is_Wednesday']= wednesday
series['is_Thursday']= thursday
series['is_Friday']= friday
```

#### Crea feature 'is_holiday' che vale 1 se Ã¨ un giorno festivo (Pasqua, Liberazione ecc..), 0 altrimenti

```python
# crea campo 'is_public_holiday'
# Pasqua: domenica 16 aprile 2017
# LunedÃ¬ dell'Angelo: lunedÃ¬ 17 aprile 2017
# Festa della Liberazione: martedÃ¬ 25 aprile 2017
# Festa del Lavoro: lunedÃ¬ 1 maggio 2017
# Festa della Repubblica: venerdi 2 giugno 2017
# Ferragosto: martedÃ¬ 15 agosto 2017
def find_holiday_days(data):
    holiday = []
    for i in data['date']:
        #print(i)
        if ((i == '2017-04-16') | (i == '2017-04-17') | (i == '2017-04-25') | 
            (i == '2017-05-01') | (i == '2017-06-02') | (i == '2017-08-15')): 
            holiday.append(1)
        else: 
            holiday.append(0)
     
    return holiday

hd = find_holiday_days(series)
series['is_holiday']= hd  
series.head(3)
```

#### Crea feature 'is_business_hour' che vale 1 se 9.00 <= osservazione <= 21.00 , 0 altrimenti

```python
def find_business_hours(data):
    business_h = []
    for i in data['hours']:
        if i > 8 and i < 21 :
            business_h.append(1)
        
        else:
            business_h.append(0)
    return business_h  

business_h = find_business_hours(series)        
series['is_business_hour'] = business_h
series.tail(3)
```

#### Costruisco i lag

```python
# split a univariate sequence into samples
def split_sequence(sequence, n_steps):
    X, y = list(), list()
    for i in range(len(sequence)):
        # find the end of this pattern
        end_ix = i + n_steps
        # check if we are beyond the sequence
        if end_ix > len(sequence)-1:
            break
        # gather input and output parts of the pattern
       
        seq_x, seq_y = sequence[i:end_ix], sequence[end_ix]
        X.append(seq_x)
        y.append(seq_y)
    return np.array(X), np.array(y)

# choose a number of time steps for the window
n_steps = 96

# data 
dati = series['nr_people'].values

# split into samples
X_lag, y_lag = split_sequence(dati, n_steps)

count = 0
# summarize the data
for i in range(len(X_lag)):
    count +=1
    if count > 2: 
        break
    print(X_lag[i], y_lag[i])    
```

```python
new_features = series[['is_holiday']]
X_new = new_features.values
```

```python
X = np.concatenate((X_lag, X_new[0:len(X_lag)]),axis=1)
print(len(X))
```

```python
y_final = y[0:len(X)]
len(y_final)
```

```python
train_X, test_X, train_y, test_y = train_test_split(X, y_final, test_size=0.2, random_state=1)
train_X, val_X, train_y, val_y = train_test_split(train_X, train_y, test_size=0.25, random_state=1)
print(len(train_X), len(train_y), len(test_X), len(test_y), len(val_X), len(val_y))
```

```python
# define model
model = Sequential()
model.add(Dense(500, activation='relu', input_dim=n_steps+1)) 
model.add(Dense(1))

model.compile(loss='mse',optimizer='adam', metrics=[metrics.mae, 'accuracy'])

model.fit(train_X, train_y, epochs=500, verbose=2)
```

```python
# compute error
predicted = np.array([])

for i in range(len(test_X)):
    x_input = test_X[i].reshape((1, n_steps+1))
    yhat = model.predict(x_input, verbose=0)
    predicted = np.append(predicted, yhat[0]) 

expected = test_y  
   
# abs difference
difference = abs((expected - predicted))
```
