### Crea nuove features

#### Crea feature 'is_buisness_day' che vale 1 se è un giorno lavorativo, 0 altrimenti  

```python
# crea un campo 'is_business_day' con valore binario
def find_business_days(data):
    business_day = []
    for i in data['date']:
        #print(i)
        s = i.split('-')
        year = int(s[0])
       # print(year, '----------------_>>>>><')
        month = int(s[1])
        day = int(s[2])
        ##print(day, ' ', month, ' ', year)
    
        dayofweek = datetime.date(year, month, day).strftime("%A")
        if ((dayofweek == 'Saturday') | (dayofweek == 'Sunday')): 
            business_day.append(0)
        else: 
            business_day.append(1)
     
    return business_day

bd = find_business_days(series)
series['is_business_day']= bd  
```

#### Crea feature 'is_April', 'is_May' ecc.. che vale 1 se si tratta del mese in questione, vale 0 altrimenti

```python
# crea un campo 'is_Month' con valore binario
def find_month(data):
    april = []
    may = []
    june = []
    july = []
    august = []
    
    for i in data['date']:
        #print(i)
        s = i.split('-')
        year = int(s[0])
       # print(year, '----------------_>>>>><')
        month = int(s[1])
        #print(month)
        if month == 4:
            april.append(1)
            may.append(0)
            june.append(0)
            july.append(0)
            august.append(0)
        if month == 5:
            april.append(0)
            may.append(1)
            june.append(0)
            july.append(0)
            august.append(0)
        if month == 6:
            april.append(0)
            may.append(0)
            june.append(1)
            july.append(0)
            august.append(0)
        if month == 7:
            april.append(0)
            may.append(0)
            june.append(0)
            july.append(1)
            august.append(0) 
        if month == 8:
            april.append(0)
            may.append(0)
            june.append(0)
            july.append(0)
            august.append(1)     
     
    return april, may, june, july, august

april, may, june, july, august = find_month(series)
series['is_April']= april
series['is_May']= may
series['is_June']= june
series['is_July']= july
series['is_August']= august
```

#### Crea feature is_Day es. 'is_Monday' che vale 1 se si tratta del giorno in questione, vale 0 altrimenti

```python
# crea un campo 'is_Day' con valore binario
def find_month(data):
    monday = []
    tuesday = []
    wednesday = []
    thursday = []
    friday = []
    
    for i in data['date']:
        s = i.split('-')
        year = int(s[0])
        month = int(s[1])
        day = int(s[2])
        
        dayofweek = datetime.date(year, month, day).strftime("%A")
        if (dayofweek == 'Monday'): 
            monday.append(1)
            tuesday.append(0)
            wednesday.append(0)
            thursday.append(0)
            friday.append(0)
        if (dayofweek == 'Tuesday'):
            monday.append(0)
            tuesday.append(1)
            wednesday.append(0)
            thursday.append(0)
            friday.append(0)
        if (dayofweek == 'Wednesday'):
            monday.append(0)
            tuesday.append(0)
            wednesday.append(1)
            thursday.append(0)
            friday.append(0)
        if (dayofweek == 'Thursday'):
            monday.append(0)
            tuesday.append(0)
            wednesday.append(0)
            thursday.append(1)
            friday.append(0) 
        if (dayofweek == 'Friday'):
            monday.append(0)
            tuesday.append(0)
            wednesday.append(0)
            thursday.append(0)
            friday.append(1)     
        if (dayofweek == 'Saturday'):
            monday.append(0)
            tuesday.append(0)
            wednesday.append(0)
            thursday.append(0)
            friday.append(0)          
        if (dayofweek == 'Sunday'):
            monday.append(0)
            tuesday.append(0)
            wednesday.append(0)
            thursday.append(0)
            friday.append(0) 
              
    return monday, tuesday, wednesday, thursday, friday

monday, tuesday, wednesday, thursday, friday = find_month(series)
print(len(monday))
print(len(tuesday))
print(len(friday))
series['is_Monday']= monday
series['is_Tuesday']= tuesday
series['is_Wednesday']= wednesday
series['is_Thursday']= thursday
series['is_Friday']= friday
```

#### Crea feature 'is_holiday' che vale 1 se è un giorno festivo (Pasqua, Liberazione ecc..), 0 altrimenti

```python
# crea campo 'is_public_holiday'
# Pasqua: domenica 16 aprile 2017
# Lunedì dell'Angelo: lunedì 17 aprile 2017
# Festa della Liberazione: martedì 25 aprile 2017
# Festa del Lavoro: lunedì 1 maggio 2017
# Festa della Repubblica: venerdi 2 giugno 2017
# Ferragosto: martedì 15 agosto 2017
def find_holiday_days(data):
    holiday = []
    for i in data['date']:
        #print(i)
        if ((i == '2017-04-16') | (i == '2017-04-17') | (i == '2017-04-25') | 
            (i == '2017-05-01') | (i == '2017-06-02') | (i == '2017-08-15')): 
            holiday.append(1)
        else: 
            holiday.append(0)
     
    return holiday

hd = find_holiday_days(series)
series['is_holiday']= hd  
series.head(3)
```

#### Crea feature 'is_business_hour' che vale 1 se 9.00 <= osservazione <= 21.00 , 0 altrimenti

```python
def find_business_hours(data):
    business_h = []
    for i in data['hours']:
        if i > 8 and i < 21 :
            business_h.append(1)
        
        else:
            business_h.append(0)
    return business_h  

business_h = find_business_hours(series)        
series['is_business_hour'] = business_h
series.tail(3)
```

```python
meteo = pd.read_csv('MeteoBologna.csv')
meteo
```

```python
meteo_binary = {}
for index, row in meteo.iterrows():
    #print(row['DATA'], row['FENOMENI'])
    date = row['DATA']
    date2 = date.split('-')
    gg = date2[0]
    mm = date2[1]
    aaaa = date2[2]
    date = aaaa+'-'+mm+'-'+gg
    osservazione = str(row['FENOMENI'])
    if 'pioggia' in osservazione:
        meteo_binary[date] = 1
    else:  
        meteo_binary[date] = 0
meteo_binary        
```

```python
rainy_shiny_days = []
for index, row in series.iterrows():
    date_i = row['date']
    meteo_bin_result = meteo_binary[date_i]
    rainy_shiny_days.append(meteo_bin_result)

series['is_rainy'] = rainy_shiny_days    
```

### Metto insieme i features creati ('is_holiday') con i lag per vedere se posso migliorare la previsione 

```python
series2 = agg_by_cell.get_group('488-1264')['nr_people'].values
```

```python
# split a univariate sequence into samples
def split_sequence(sequence, n_steps):
    X, y = list(), list()
    for i in range(len(sequence)):
        # find the end of this pattern
        end_ix = i + n_steps
        # check if we are beyond the sequence
        if end_ix > len(sequence)-1:
            break
        # gather input and output parts of the pattern
       
        seq_x, seq_y = sequence[i:end_ix], sequence[end_ix]
        X.append(seq_x)
        y.append(seq_y)
    return np.array(X), np.array(y)
```

```python
# choose a number of time steps for the window
n_steps = 96

# split into samples
X_lag, y_lag = split_sequence(series2, n_steps)

count = 0
# summarize the data
for i in range(len(X_lag)):
    print(X_lag[i], y_lag[i])
    count +=1
    if count > 1: 
        break
```
