```python
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error
from math import sqrt
from statsmodels.tsa.ar_model import AR
import warnings
from statsmodels.tsa.vector_ar.var_model import VAR
from random import random
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('/Users/alket/Desktop/dati/new_data_backfill_forwfill.csv',index_col = 0)
```

```python
agg_by_cell = data.groupby(by = data['cell_num'])
```

```python
num_data = []
## prepare dataset
for i, k in agg_by_cell:
    cell_i = agg_by_cell.get_group(i)
    #print(cell_i['nr_people'][:1])
    # define input sequence
    series_i = cell_i['nr_people'].values
    series_i = series_i.reshape((len(series_i), 1))
    num_data.append(series_i)
```

```python
print(len(num_data))
print(type(num_data[0]))
```

```python
final_data = np.hstack((num_data))
print(type(final_data))
print(len(final_data))
print('shape = ', final_data.shape)

print(final_data[0][5])
```

```python
#for i in range(96):
#    print(final_data[i][0])
```

```python
split_train_test = int(len(final_data)*0.75)
train_X, test_X = final_data[:split_train_test], final_data[split_train_test:]
train_y, test_y = final_data[:split_train_test], final_data[split_train_test:]
print(len(train_X), len(train_y), len(test_X), len(test_y))
```

```python
model = VAR(train_X)
model_fit = model.fit()
```

```python
## set number of steps
n_steps = 5
```

```python
yhat = model_fit.forecast(test_y, steps=n_steps)
print(yhat)
```

```python
yhat.shape
```

```python
predicted = []
expected = np.array(test_y)
expected.shape

for i in range(len(test_X)):
    yhat = model_fit.forecast(test_y, steps=n_steps)
    predicted.append(np.rint(yhat[0]))
    
predicted = np.asarray(predicted)
print(predicted[0:6])
print(expected[0:6])  
```

```python
print(predicted.shape)
print(expected.shape)
```

```python
## boxlot 
d = abs(predicted-expected)
print(d.shape)

plt.figure(figsize = (14,32))
plt.boxplot(d, vert= False);
plt.show()
```
