```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
from random import random
import matplotlib.pyplot as plt 
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
import pandas as pd
import numpy as np
from pandas import datetime
from random import random
from pandas.plotting import lag_plot
from pandas.plotting import autocorrelation_plot
from statsmodels.graphics.tsaplots import plot_acf
from sklearn.metrics import mean_squared_error
from math import sqrt
from statsmodels.tsa.ar_model import AR
import warnings
warnings.filterwarnings('ignore')
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
data = pd.read_csv('/Users/alket/Desktop/dati/new_data_backfill_forwfill.csv',index_col = 0)
```

```python
gbc = data.groupby(by = data['cell_num'])
cell_1 = gbc.get_group('488-1264')
series1 = cell_1['nr_people']

smoothing_level=0.2
```

```python
fit1 = SimpleExpSmoothing(series1).fit(smoothing_level,optimized=True)
fcast1 = fit1.forecast(len(series1)).rename('alpha = '+str(smoothing_level))

ax = series1.plot(marker='o', color='black', figsize=(18,10))
fcast1.plot(marker='o', ax=ax, color='blue', legend=True)
fit1.fittedvalues.plot(marker='o', ax=ax, color='blue')
```

```python
series = np.array(series1)
predictions = np.array(fcast1)

error = abs(series - predictions)
print('Mean Abs Error ', error.mean())
plt.figure(figsize = (14,1))
plt.boxplot(error, vert= False); 
```

### Calcolo per tutte le celle 

```python
group_data_by_cell = data.groupby(by = data['cell_num'])
smoothing_level=0.01
dict2data = {}
dict2MAPE = {}
for i, k in group_data_by_cell:
    
    cell_data_i = k
    cell = i
    series_i = cell_data_i['nr_people']
    train, test = series_i[:9000], series_i[9000:]
    
    fit1 = SimpleExpSmoothing(train).fit(smoothing_level,optimized=True)
    fcast1 = fit1.forecast(len(test)).rename('alpha = '+str(smoothing_level))
    len(fcast1) 
    expected = np.array(test)
    predictions = np.array(fcast1)
    
    error = abs(expected - predictions)
    print('Mean Abs Error ', error.mean())
    print(cell)
    plt.figure(figsize = (14,1))
    plt.boxplot(error, vert= False);  
    plt.show()
    print('Cell distribution :')
    plt.figure(figsize = (14,1))
    plt.boxplot(series_i, vert= False);  
    plt.show()
    
    MAPE = np.mean(abs(100 * (error/expected)))
    dict2MAPE[cell] = MAPE
    
    # collect data 2 dictionary
    minimum = np.amin(error)   
    per75 = np.percentile(error, 75)
    per50 = np.percentile(error, 50)
    per25 = np.percentile(error, 25)
    maximum = np.amax(error)
    l5i = [minimum, per25, per50, per75, maximum]
    dict2data[cell] = l5i
```

### Scrivi i dati

```python
with open('MAE_error_data_4_exponential_smoothing_0.01.csv', 'w') as f:
    for key, value in dict2data.items():
        f.write('%s:%s\n' % (key, value))
        
with open('MAPE_error_data_4_exponential_smoothing_0.01.csv', 'w') as f:
    for key, value in dict2MAPE.items():
        f.write('%s:%s\n' % (key, value))        
```
