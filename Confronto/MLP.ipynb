### Prepare input data for MLP model

```python
# split a univariate sequence into samples
def split_sequence(sequence, n_steps_in, n_steps_out):
    X, y = list(), list()
    for i in range(len(sequence)):
         # find the end of this pattern
         end_ix = i + n_steps_in
         out_end_ix = end_ix + n_steps_out
         # check if we are beyond the sequence
         if out_end_ix > len(sequence):
             break
         # gather input and output parts of the pattern
         seq_x, seq_y = sequence[i:end_ix], sequence[end_ix:out_end_ix]
         X.append(seq_x)
         y.append(seq_y)
    return np.array(X), np.array(y)

# choose a number of time steps
n_steps_in, n_steps_out = 96,22

# split into samples
X, y = split_sequence(series, n_steps_in, n_steps_out)
count = 0
for i in range(len(X)):
    print(series[i],X[i], y[i])
    count +=1
    if count>1: break
```

### Prepare train -test set

```python
train_size = 8000

train_X, test_X = X[:train_size], X[train_size:]
train_y, test_y= y[:train_size], y[train_size:]
print(train_X.shape, test_X.shape)
```

### Define MLP model 

```python
# define model
model = Sequential()
model.add(Dense(500, activation='relu', input_dim=n_steps_in)) 
model.add(Dense(n_steps_out))

#opt = SGD(lr=0.01, momentum=0.9)
model.compile(optimizer='adam', loss='mse' , metrics=[metrics.mae, 'accuracy'])

# fit model
model.fit(train_X, train_y, batch_size=64, epochs=200, verbose=2)
```

### Compute forecasting error - MAE (Mean Absolute Error)

```python
# predict and compute error
expected = test_y
predicted = []
for i in range(len(test_X)):
    x_input = test_X[i].reshape((1, n_steps_in)) # + 1 per la nuova feature aggiunta
    yhat = model.predict(x_input, verbose=0)
    #predicted.append(np.rint(yhat[0]))   
    predicted.append(np.around(yhat[0], decimals=1)) 
```

### Visualize distribution of MAE per forecasting step

```python
difference = abs(expected - predicted)
print(type(difference))
print(difference.shape[0], difference.shape[1])

# comment line below to see all the boxplots for every number of steps from 1 to n_steps_out
#difference =  np.reshape(difference, difference.shape[0] * difference.shape[1])

plt.figure(figsize = (14,8))
plt.boxplot(difference, vert= False);
plt.show()
```
