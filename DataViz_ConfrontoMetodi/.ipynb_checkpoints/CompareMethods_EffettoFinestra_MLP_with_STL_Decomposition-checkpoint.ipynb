```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
colnames = ['min', 'Q1', 'mediana', 'Q3', 'max']

# cambiare i per corsi con i propri dati
in54_25 = pd.read_csv('/Users/alket/Desktop/poliseye_timeseries/DeepLearningModels/univariate-multisteps/error_data_4_MLP_with_STL_Decomposition_54_25.csv', names=colnames, header=None)

in96_25 = pd.read_csv('/Users/alket/Desktop/poliseye_timeseries/DeepLearningModels/univariate-multisteps/error_data_4_MLP_with_STL_Decomposition_96_25.csv', names=colnames, header=None)

in25_25 = pd.read_csv('/Users/alket/Desktop/poliseye_timeseries/DeepLearningModels/univariate-multisteps/error_data_4_MLP_with_STL_Decomposition_25.csv', names=colnames, header=None)

in75_25 = pd.read_csv('/Users/alket/Desktop/poliseye_timeseries/DeepLearningModels/univariate-multisteps/error_data_4_MLP_with_STL_Decomposition_75_25.csv', names=colnames, header=None)
```

```python
max_in25_25= []
for i, row in in25_25.iterrows():
    max_i= row['max']
    max_i = max_i[0:-2]
    max_in25_25.append(float(max_i))
    
max_in54_25= []
for i, row in in54_25.iterrows():
    max_i= row['max']
    max_i = max_i[0:-1]
    max_in54_25.append(float(max_i)) 
    
max_in96_25= []
for i, row in in96_25.iterrows():
    max_i= row['max']
    max_i = max_i[0:-1]
    max_in96_25.append(float(max_i))   
    
max_in75_25= []
for i, row in in75_25.iterrows():
    max_i= row['max']
    max_i = max_i[0:-1]
    max_in75_25.append(float(max_i))     
```

```python
max_in25_25 = np.array(max_in25_25)
max_in54_25 = np.array(max_in54_25)
max_in96_25 = np.array(max_in96_25)
max_in75_25 = np.array(max_in75_25)
```

```python
med_in25_25 = np.array(in25_25['mediana'].tolist())
med_in54_25 = np.array(in54_25['mediana'].tolist())
med_in96_25 = np.array(in54_25['mediana'].tolist())
med_in75_25 = np.array(in54_25['mediana'].tolist())
```

```python
mediana_plot = [med_in25_25, med_in54_25, med_in75_25, med_in96_25]

# Create a figure instance
fig = plt.figure(1, figsize=(20, 7))

# Create an axes instance
ax = fig.add_subplot(111)

# Create the boxplot
bp = ax.boxplot(mediana_plot, patch_artist=True)

ax.set_xticklabels(['25-25', '54-25', '75_25', '96-25'])


# Save the figure
#fig.savefig('ConfrontoMetodi.jpeg', bbox_inches='tight')
```
