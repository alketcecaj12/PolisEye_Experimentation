### Compare methods 

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

### Compare methods by MAE

```python
colnames = ['min', 'Q1', 'mediana', 'Q3', 'max']

# cambiare i percorsi con i propri dati
data_Prophet = pd.read_csv('/Users/alket/Desktop/poliseye_timeseries/Prophet/'+
                 'error_data_4_Prophet_prediction.csv', names=colnames, header=None)

data_MLP = pd.read_csv('/Users/alket/Desktop/poliseye_timeseries/DeepLearningModels/univariate-multisteps/'+
                 'error_data_4_MLP_with_STL_Decomposition_in24_out25_period_672.csv', names=colnames, header=None)

data_LSTM = pd.read_csv('/Users/alket/Desktop/poliseye_timeseries/DeepLearningModels/univariate-multisteps/'+
                 'error_data_4_MLP_with_STL_Decomposition_in24_out25_period_672.csv', names=colnames, header=None)

data_CNN = pd.read_csv('/Users/alket/Desktop/poliseye_timeseries/DeepLearningModels/univariate-multisteps/'+
                 'error_data_4_CNN_with_STL_Decomposition_in96_out38_period_96.csv', names=colnames, header=None)

data_ARIMA = pd.read_csv('/Users/alket/Desktop/poliseye_timeseries/DeepLearningModels/univariate-multisteps/'+
                 'error_data_4_CNN_with_STL_Decomposition_in96_out38_period_96.csv', names=colnames, header=None)

data_SARIMA = pd.read_csv('/Users/alket/Desktop/poliseye_timeseries/DeepLearningModels/univariate-multisteps/'+
                 'error_data_4_CNN_with_STL_Decomposition_in96_out38_period_96.csv', names=colnames, header=None)

data_ETS = pd.read_csv('/Users/alket/Desktop/poliseye_timeseries/DeepLearningModels/univariate-multisteps/'+
                 'error_data_4_CNN_with_STL_Decomposition_in96_out38_period_96.csv', names=colnames, header=None)

data_AR_deco = pd.read_csv('/Users/alket/Desktop/poliseye_timeseries/StatisticalModels/UnivariateSeriesPrediction/'+
                 'error_data_4_Autoregression_with_STL_Decom__37_StepForecast.csv', names=colnames, header=None)
```

```python
mae_MLP = np.array(data_MLP['mediana'].tolist())
mae_Prophet = np.array(data_Prophet['mediana'].tolist())
mae_ETS = np.array(data_ETS['mediana'].tolist())
mae_CNN = np.array(data_CNN['mediana'].tolist())
mae_AR = np.array(data_AR_deco['mediana'].tolist())
```

### Mediana errore per ognbi metodo 

```python
data_to_plot = [mae_CNN, mae_MLP, mae_Prophet, mae_AR, mae_ETS]

# Create a figure instance
fig = plt.figure(1, figsize=(20, 7))

# Create an axes instance
ax = fig.add_subplot(111)

# Create the boxplot
bp = ax.boxplot(data_to_plot, patch_artist=True)
ax.set_xticklabels(['CNN', 'MLP', 'Prophet', 'AR', 'mae_AR_deco'])

# Save the figure
fig.savefig('ConfrontoMetodi.jpeg', bbox_inches='tight')
```

```python
print(np.mean(mae_MLP))
np.mean(mae_AR_deco)
```

### Max errore per ogni metodo 

```python
max_AR= []
for i, row in data_AR.iterrows():
    max_i= row['max']
    max_i = max_i[0:-2]
    max_AR.append(float(max_i))

    
max_AR_deco = []
for i, row in data_AR_deco.iterrows():
    max_i= row['max']
    max_i = max_i[0:-2]
    max_AR_deco.append(float(max_i))

max_MLP= []
for i, row in data_MLP.iterrows():
    max_i= row['max']
    max_i = max_i[0:-1]
    max_MLP.append(float(max_i))
    
max_Prophet= []
for i, row in data_Prophet.iterrows():
    max_i= row['max']
    max_i = max_i[0:-1]
    max_Prophet.append(float(max_i)) 
    
max_CNN= []
for i, row in data_CNN.iterrows():
    max_i= row['max']
    max_i = max_i[0:-1]
    max_CNN.append(float(max_i))     

max_CNN = np.array(max_CNN)    
max_MLP = np.array(max_MLP)
max_Prophet = np.array(max_Prophet)
max_AR = np.array(max_AR)
max_AR_deco = np.array(max_AR_deco)
```

```python
data_to_plot = [max_CNN, max_MLP, max_Prophet, max_AR, max_AR_deco]

# Create a figure instance
fig = plt.figure(1, figsize=(20, 7))

# Create an axes instance
ax = fig.add_subplot(111)

# Create the boxplot
bp = ax.boxplot(data_to_plot, patch_artist=True)
ax.set_xticklabels(['CNN', 'MLP', 'Prophet', 'AR', 'AR_deco'])

# Save the figure
fig.savefig('ConfrontoMetodi.jpeg', bbox_inches='tight')
```
